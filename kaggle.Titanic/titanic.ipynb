```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeClassifier

from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.metrics import precision_score, recall_score
```

```python
gender_submission = pd.read_csv('gender_submission.csv')
train = pd.read_csv('train.csv')
subm = pd.read_csv('test.csv')
```

```python
# Drop not informative features
X = train.drop(['PassengerId', 'Survived', 'Name', 'Ticket', 'Cabin'], axis=1)
y = train.Survived
# Convert string categorial var into numeric
X = pd.get_dummies(X)
```

```python
# Fill NA in Age column

# Using median
# X = X.fillna({'Age': X.Age.median()})

# using LinearRegression
def fill_na_using_linear_regression(df, column):
    X_train = df[df[column].notna()].drop(column, axis=1)
    y_train = df[df[column].notna()][column]
    X_test = df[df[column].isna()].drop(column, axis=1)
    lr = LinearRegression()
    lr.fit(X_train, y_train)
    #df[df[column].isna()][column] = pd.Series(lr.predict(X_test))
    predicted = pd.Series(lr.predict(X_test))
    df[column].fillna(predicted, inplace=True)
    #print()
    return df

X = fill_na_using_linear_regression(X, 'Age')
X[X.Age.isna()]
```

```python
S = subm.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1)
S = pd.get_dummies(S)
S = S.fillna({'Fare': S.Fare.median()})
S = fill_na_using_linear_regression(S, 'Age')
```

```python
# Split dataset into test and train
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
dtc = DecisionTreeClassifier()
params = {'criterion': ('gini', 'entropy'), 'max_depth': range(1, 30)}
search = GridSearchCV(dtc, params, cv=5)
search.fit(X, y)
```

```python
recall_score(y_test, search.best_estimator_.predict(X_test))
```

```python
submission = search.predict(S)
submission = pd.DataFrame({'PassengerId': subm['PassengerId'], 'Survived': submission})
submission.to_csv('submission.csv', index=False)
```
