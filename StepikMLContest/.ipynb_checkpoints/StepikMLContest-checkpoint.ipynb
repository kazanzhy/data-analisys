```python
import pandas as pd
import numpy as np
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(rc={'figure.figsize': (9, 6)})
```

```python
# Define global vars
path = '~/git/data-analisys/stepik.MLContest/'
```

```python
# Load data
events_data = pd.read_csv(path + 'event_data_train.zip', compression='zip')
submissions_data = pd.read_csv(path + 'submissions_data_train.zip', compression='zip')

#Converting timestamp to datetime and adding day
events_data['date'] = pd.to_datetime(events_data.timestamp, unit='s')
events_data['day'] = events_data.date.dt.date
submissions_data['date'] = pd.to_datetime(submissions_data.timestamp, unit='s')
submissions_data['day'] = submissions_data.date.dt.date
```

```python
# Building plot of users' activity grouped by day
events_data.groupby('day').user_id.nunique().plot()
```

```python
# Calculate count of users actions
users_events_data = events_data.pivot_table(index='user_id', 
                                        columns='action',
                                        values='step_id', 
                                        aggfunc='count', 
                                        fill_value=0).reset_index().sort_values('viewed', ascending=False)
users_events_data.head()
```

```python
# Calculate count of correct and wrong answers
users_scores = submissions_data.pivot_table(index='user_id', 
                        columns='submission_status', 
                        values='step_id', 
                        aggfunc='count', 
                        fill_value=0).reset_index()
users_scores.head()
```

```python
# Calculate breaks between steps for each user
gap_data = events_data[['user_id', 'day', 'timestamp']]\
        .drop_duplicates(subset=['user_id', 'day'])\
        .groupby('user_id')['timestamp']\
        .apply(list)\
        .apply(np.diff).values
# Convert array of arrays to Series
gap_data = pd.Series(np.concatenate(gap_data, axis=0))
# Convert seconds to days
gap_data = gap_data / (24 * 60 * 60)
# Duration of break (in days) for returning of 92.5% users 
gap_data.quantile(0.925) # 59.760440972222156
```

```python
# Last second of  the data i.e. current time
now = int(events_data.tail(1).timestamp)
drop_out_threshold = 30 * 24 * 60 * 60
# When user was on course last time
users_data = events_data.groupby('user_id', as_index=False)\
        .agg({'timestamp': 'max'})\
        .rename(columns={'timestamp': 'last_timestamp'})
# Was user online last month
users_data['is_gone_user'] = (now - users_data.last_timestamp) > drop_out_threshold
users_data.head()
```

```python
# Merge data about correct and wrong answers
users_data = users_data.merge(users_scores, how='outer').fillna(0)
# Merge data about count of actions
users_data = users_data.merge(users_events_data, how='outer')
# Add info about count days in the course
users_days = events_data.groupby('user_id', as_index=False).agg({'day': 'nunique'})
users_data = users_data.merge(users_days, how='outer')
# Add column of success passed course
users_data['passed_course'] = users_data.passed > 170
```

```python
user_min_time = events_data.groupby('user_id', as_index=False)\
    .agg({'timestamp': 'min'})\
    .rename(columns={'timestamp': 'min_timestamp'})
users_data = users_data.merge(user_min_time, how='outer')
```

```python
# Select first 3 days data
threshold = 3 * 24 * 60 * 60
events_data_train = events_data.merge(user_min_time, how='outer').query('timestamp - min_timestamp <= @threshold')
submissions_data_train = submissions_data.merge(user_min_time, how='outer').query('timestamp - min_timestamp <= @threshold')
```

```python
# Make train set
# How many days user was on course
X = submissions_data_train.groupby('user_id').day.nunique().to_frame().reset_index().rename(columns={'day': 'days'})
# How many submissoins user tried to pass
step_tried = submissions_data_train.groupby('user_id').step_id.nunique().to_frame().reset_index().rename(columns={'step_id': 'step_tried'})
X = X.merge(step_tried, on='user_id', how='outer')
# Final submissions' scores 
scores = submissions_data_train.pivot_table(index='user_id', columns='submission_status', values='step_id', aggfunc='count', fill_value=0).reset_index()
X = X.merge(scores, on='user_id', how='outer')
X['correct_ratio'] = X.correct / (X.correct + X.wrong)
# How many steps user viewed
actions = events_data_train.pivot_table(index='user_id', columns='action', values='step_id', aggfunc='count', fill_value=0).reset_index()[['user_id', 'viewed']]
X = X.merge(actions, on='user_id', how='outer')
X = X.fillna(0)
# Add info about passing course
X = X.merge(users_data[['user_id', 'passed_course', 'is_gone_user']], on='user_id', how='outer')
# 
X = X[-((X.is_gone_user == False) & (X.passed_course == False))]
y = X.passed_course.map(int)
X = X.set_index(X.user_id)
# Drop unused columns
X = X.drop(['user_id', 'passed_course', 'is_gone_user'], axis=1)
X.head()
```
